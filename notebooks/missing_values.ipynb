```python
# Install required packages (only needed once in Colab)
!pip install scikit-learn pandas

import pandas as pd
from sklearn.datasets import load_diabetes

# Load the Diabetes dataset
diabetes = load_diabetes(as_frame=True)

# Create DataFrame with features + target column
df = diabetes.frame.copy()
df["target"] = diabetes.target

# Output 1: Dataset shape and preview
print("Output 1: Dataset Shape and Preview")
print("Shape (rows, columns):", df.shape)
print("First 5 rows:")
display(df.head())

# Output 2: Missing values check
print("\nOutput 2: Missing Values Analysis")
missing_pct = (df.isna().sum() / len(df) * 100).round(2)
print("Percentage of missing values by column:")
display(missing_pct.rename("Missing (%)").to_frame())

# Summary
if missing_pct.sum() == 0:
    print("\nNo missing values found in this dataset.")
else:
    print("\nSome missing values were found, check the table above.")
```
